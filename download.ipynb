Funzione in grado di procedere ad analizzare una pagina di una search del sito arxiv.org sfruttando xpath.
La seguente funzione procede ad andare a cercare i link che portano all' anteprima di ogni paper presente, ottenendo cosÃ¬ un nuovo link, questo link va modificato sostituendo la 'x' con il 5, in questo modo si viene reindirizzati al sito ausiliario ar5iv.labs.arxiv.org.

Esempio:
Se ci troviamo nella pagina: https://arxiv.org/abs/2407.16234
riscrivendolo come https://ar5iv.org/abs/2407.16234 ed effettuando la sua request, si viene reindirizzati alla pagina html: https://ar5iv.labs.arxiv.org/html/2407.16234

da qui si puÃ² procedere a scaricare il singolo paper in formato mhtml, la sua relativa conversione html avverrÃ  successivamente.

```python
def estrai_link_papers(html_content):
  """Estrae i link delle pagine HTML dei singoli papers da una pagina di ricerca di arXiv.org.

  Args:
    html_content: Il contenuto HTML della pagina di ricerca.

  Returns:
    Una lista di tuple contenenti l'URL modificato e il codice del documento.
  """

  soup = BeautifulSoup(html_content, 'html.parser')

  link_papers = []
  for link in soup.select("ol.breathe-horizontal li div p a[href*='abs']"): 
    paper_url = link['href']

    # Sostituisci "x" con "5" per ottenere l'URL di ar5iv
    ar5iv_url = paper_url.replace("x", "5")

    # Estrai il codice del documento
    codice_documento = paper_url.split('/')[-1]

    # Aggiungi il link modificato e il codice del documento alla lista
    link_papers.append((ar5iv_url, codice_documento))

  return link_papers

#Creo il file txt dove salverÃ² i link se non esiste giÃ 
if not os.path.exists("links.txt"):
    open("links.txt", "w").close()
    
# Esempio di utilizzo
for page in os.listdir("searchPages"):
    with open("searchPages/" + page, "r", encoding="utf-8") as file:
        html_content = file.read()
    link_papers = estrai_link_papers(html_content)


    # Salva i link in un file di testo (in modalitÃ  "a" per appendere)
    with open("links.txt", "a", encoding="utf-8") as file:
        for ar5iv_url, codice_documento in link_papers:
            file.write(f"Link: {ar5iv_url}\n")
            file.write(f"Codice: {codice_documento}\n")
```

Ora procediamo a scaricare tutti i papers in html leggendo i links presenti in readme

```python
def download_paper_html(link, codice):
  """Scarica la pagina HTML associata ad un link e la salva nella cartella 'sources'.

  Args:
    link: L'URL del paper.
    codice: Il codice del documento.
  """

  try:
    print(f"Scaricando pagina {codice}: {link}")
    response = requests.get(link)
    response.raise_for_status()  # Gestisci gli errori di richiesta HTTP
    
    # Crea la cartella 'sources' se non esiste
    os.makedirs("sources", exist_ok=True)

    # Salva la pagina HTML nel file con nome pari al codice
    with open(f"sources/{codice}.html", "w", encoding="utf-8") as file:
      file.write(response.text)

    print(f"Pagina {codice} scaricata correttamente.")

  except requests.exceptions.RequestException as e:
    print(f"Errore durante il download di {codice}: {e}")

# Esempio di utilizzo
with open("links.txt", "r", encoding="utf-8") as file:
  for line in file:
    if line.startswith("Link:"):
      link = line.strip().split(" ")[1]
    elif line.startswith("Codice:"):
      codice = line.strip().split(" ")[1]
      download_paper_html(link, codice)

print("Download Completed.")
```

```python
html_pages = os.listdir("sources")

for page in html_pages:
    with open("sources/" + page, "r", encoding="utf-8") as file:
        html_content = file.read()

    soup = BeautifulSoup(html_content, 'html.parser')

    if soup.find("a",class_="is-sr-only") != None:
        os.remove(f"sources/{page}")
```
